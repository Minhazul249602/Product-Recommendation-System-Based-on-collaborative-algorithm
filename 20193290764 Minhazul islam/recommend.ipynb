```python
number_rating_df=ratings_wit_name.groupby('title').count()['average_rating'].reset_index()
number_rating_df.rename(columns={'average_rating':'num_ratings'},inplace=True)
number_rating_df
```

```python
# Convert 'average_rating' column to numeric type
ratings_wit_name['average_rating'] = pd.to_numeric(ratings_wit_name['average_rating'], errors='coerce')

# Drop rows with missing values in 'average_rating' column
ratings_wit_name.dropna(subset=['average_rating'], inplace=True)

# Group by 'title' and calculate mean of 'average_rating'
avg_rating_df = ratings_wit_name.groupby('title')['average_rating'].mean().reset_index()

# Rename the column
avg_rating_df.rename(columns={'average_rating': 'avg_ratings'}, inplace=True)

# Display the resulting DataFrame
print(avg_rating_df)
```

```python
popular_df=number_rating_df.merge(avg_rating_df,on='title')
popular_df
```

##Collaborating Filtering##Collaborating Filtering


```python
x=ratings_wit_name.groupby('user_id').count()['rating']>100
readers_users=x[x].index
```

```python
filtered_rating=ratings_wit_name[ratings_wit_name['user_id'].isin(readers_users)]
```

```python
y=filtered_rating.groupby('title').count()['rating']>=50
famous_books=y[y].index
```

```python
final_ratings=filtered_rating[filtered_rating['title'].isin(famous_books)]
```

```python
pt=final_ratings.pivot_table(index='title',columns='user_id',values='rating',fill_value=0)
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores=cosine_similarity(pt)
```

```python
sorted(list(enumerate(similarity_scores[2])),key=lambda x:x[1],reverse=True)
```

```python
def recommend(book_name):
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('title')['title'].values))
        #item.extend(list(temp_df.drop_duplicates('title')['authors'].values))
        #item.extend(list(temp_df.drop_duplicates('title')['image_url'].values))
        data.append(item)
    
    return data
```

```python
recommend('Jane Eyre')
```
